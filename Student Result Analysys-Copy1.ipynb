```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Student_Scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

#Drop unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

#Change Weekly Study Hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

#Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x="Gender",color="green")
ax=ax.bar_label(ax.containers[0])
#sns.countplot(data=df, x="Gender", palette="pastel",)
#sns.countplot(data=df, x="Gender", palette="Set1")
plt.title("Gender Distribution",color="red")
plt.show()
```

#In the above chart we have analysed that:
#the number females in the data is more than the number of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb,annot=True)
# plt.title("Relationship between Parent's Eduction and Student's Score")
plt.show()
```

#in the above chart we have concluded that the education of the parents have good impact
#on their Student's score

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.show()
```

#in the above chart we have analysed  that the Marital Status of the parents have no impact on their student's score

#Distribution of Ethnic groups 

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()
# print(groupA["EthnicGroup"])
# print(groupB["EthnicGroup"])
# print(groupC["EthnicGroup"])
# print(groupD["EthnicGroup"])
# print(groupE["EthnicGroup"])

l=['group A', 'group B', 'group C', 'group D', 'group E']
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
# figcolors=['orange', 'coral', 'skyblue', 'green']
#wedgeprops_largest={'width': 0.8, 'edgecolor': 'black'}

plt.figure(figsize=(3,3))
plt.pie(mylist,labels=l,autopct="%1.2f%%",startangle=140) 
plt.title("Distribution of Ethnic Groups")
plt.show()
```
